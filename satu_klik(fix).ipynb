#1girl, smile, \<lora:doll_japan:0.2\> \<lora:gi_xiangling:0.8\>

#1girl, smile, \<lora:doll_japan:0.7\> \<lora:gi_xiangling:0.7\>

0.7 + 0.7 = 1.4 / 140%
